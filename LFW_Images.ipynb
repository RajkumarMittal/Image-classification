```python
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn import ensemble
import time
```

```python
lfw = datasets.fetch_lfw_people(min_faces_per_person=100, resize=0.4)
```

```python
lfw.keys()
```

```python
lfw.data.shape
```

```python
lfw.images.shape
# Image size is 50 * 37 so image is rectangle
```

```python
fig = plt.figure(figsize=(8,8))
for i in range(64):
    ax = fig.add_subplot(8, 8, i+1)
    ax.imshow(lfw.images[i], cmap= plt.cm.bone)
plt.show()
```

```python
x, y = lfw.data, lfw.target
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)
pca_oliv = PCA()
pca_oliv.fit(x_train)
```

```python
pca_oliv.components_.shape
```

```python
k = 0
total = sum(pca_oliv.explained_variance_)
current = 0
while current/total < 0.99:
    current += pca_oliv.explained_variance_[k]
    k += 1
k
```

```python
# pca_lfw = PCA(n_components=k, whiten=True)
pca_lfw = PCA(n_components=k, whiten=False)
x_transformed = pca_lfw.fit_transform(x_train)
x_transformed.shape
```

```python
x_approx = pca_lfw.inverse_transform(x_transformed)
x_approx = x_approx.reshape((855, 50, 37))
```

```python
fig = plt.figure(figsize=(8,8))
for i in range(64):
    ax = fig.add_subplot(8, 8, i+1)
    ax.imshow(x_approx[i], cmap= plt.cm.bone)
plt.show()
```

```python
eigenv = pca_lfw.components_
eigenv.shape
```

```python
eigenface = pca_lfw.components_.reshape(k, 50, 37)
```

```python
fig = plt.figure(figsize=(8,8))
for i in range(64):
    ax = fig.add_subplot(8, 8, i+1)
    ax.imshow(eigenface[i], cmap= plt.cm.bone)
plt.show()
```

```python
x_train_pca = x_transformed
x_test_pca = pca_lfw.transform(x_test)
```

```python
rf = ensemble.RandomForestClassifier()
start = time.time()
rf.fit(x_train, y_train)
print(time.time() - start)
y_pred = rf.predict(x_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
rf = ensemble.RandomForestClassifier()
start = time.time()
rf.fit(x_train_pca, y_train)
print(time.time() - start)
y_pred = rf.predict(x_test_pca)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```
